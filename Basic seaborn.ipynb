```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
team = pd.read_csv('nba_2017_att_val_elo_with_cluster.csv')
team.head()
```

### Plot one var 4 ways

```python
plt.hist(x=team['AVG'], bins=15,color='orange')
plt.show()
```

```python
team['AVG'].hist(bins=15)
```

```python
team.plot.line(x=team.index, y='AVG')
```

### 2) plot 2 vars 3 ways
